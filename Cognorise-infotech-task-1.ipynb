```python
df = df.drop(df[df["Age"] > 90].index)
df.reset_index(drop=True, inplace=True)
df["Age"].agg(["min", "max"])
```

```python
# How much spending
```

```python
df['spending']=df['MntWines']+df['MntFruits']+df['MntMeatProducts']+df['MntFishProducts']+df['MntSweetProducts']+df['MntGoldProds']
```

```python
# Renaming The columns for better understanding of fields
```

```python
df = df.rename(columns={'MntWines' : "Wines", 'MntFruits' : 'Fruits', 'MntMeatProducts' : 'Meat', 'MntFishProducts' : 'Fish','MntSweetProducts' : 'Sweets', 'MntGoldProds' : 'Gold'})
df = df[['Age', 'Education', 'Marital_Status','Income','spending','Wines','Fruits','Meat','Fish','Sweets','Gold']]
df.head()
```

```python
# Income V/S Spending
```

```python
fig, ax =plt.subplots(1,2,figsize=(8,5))
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')
sns.kdeplot(data=df, x="Income",fill=True, ax=ax[0]).set_title(
'Income Distribution')

sns.kdeplot(data=df, x="spending",fill=True, ax=ax[1]).set_title(
'Spending Distribution')

#add space
fig.tight_layout()

fig.show() 
```
